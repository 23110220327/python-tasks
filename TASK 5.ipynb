```python
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Define data
years = ['2001', '2003', '2005', '2007', '2009']
sachin_runs = [1200, 1100, 1300, 1000, 1250]
kohli_runs = [0, 0, 600, 1050, 1200]  
sehwag_runs = [1000, 1150, 1100, 950, 900]

# Step 2: Set bar positions
x = np.arange(len(years))
width = 0.25

# Step 3: Plot side-by-side bars
plt.figure(figsize=(10, 6))
plt.bar(x - width, sachin_runs, width, label='Sachin')
plt.bar(x, kohli_runs, width, label='Kohli')
plt.bar(x + width, sehwag_runs, width, label='Sehwag')

# Step 4: Labels and legend
plt.xlabel('Year')
plt.ylabel('Runs')
plt.title('Runs Comparison: Sachin vs Kohli vs Sehwag')
plt.xticks(x, years)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Step 1: Players and their debut 5-year runs
players = ['Sachin', 'Kohli', 'Sehwag']
runs_debut_5yrs = [6000, 5700, 5300]  

# Step 2: Plot horizontal bar chart
plt.figure(figsize=(8, 5))
plt.barh(players, runs_debut_5yrs, color='orange')
plt.xlabel('Total Runs')
plt.title('Total Runs in Debut 5 Years')
plt.grid(True)
plt.show()
```

```python
# Step 1: Define labels and data
labels = ['Android', 'iOS', 'Others']
market_share = [70, 25, 5]

# Step 2: Plot pie chart
plt.figure(figsize=(6, 6))
plt.pie(market_share, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Mobile OS Market Share')
plt.show()
```

```python
# Step 1: Plot horizontal bar chart
plt.figure(figsize=(8, 4))
plt.barh(labels, market_share, color=['green', 'blue', 'gray'])
plt.xlabel('Market Share (%)')
plt.title('Mobile OS Market Share (Bar Chart)')
plt.grid(True)
plt.show()
```
