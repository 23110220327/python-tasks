```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

#  matplotlib styles
styles = [
    'ggplot', 'seaborn-v0_8', 'bmh', 'classic', 'fivethirtyeight',
    'dark_background', 'Solarize_Light2', 'grayscale',
    'seaborn-v0_8-darkgrid', 'seaborn-v0_8-whitegrid'
]

# Plot using each style
for style in styles:
    plt.style.use(style)
    plt.figure()
    plt.plot(x, y, marker='o')
    plt.title(f"Matplotlib Style: {style}")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.grid(True)
    plt.show()
```
