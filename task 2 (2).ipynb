```python
# Demonstrating multiple operations on a list

# Create a list
fruits = ["apple", "banana", "cherry", "date", "banana"]
print("Original List:", fruits)

# 1. Append an element
fruits.append("elderberry")
print("After Append:", fruits)

# 2. Insert at a specific position
fruits.insert(2, "fig")
print("After Insert at index 2:", fruits)

# 3. Remove an element by value
fruits.remove("banana")  # removes first occurrence
print("After Removing 'banana':", fruits)

# 4. Pop an element by index
popped = fruits.pop(3)
print(f"Popped element at index 3: {popped}")
print("After Pop:", fruits)

# 5. Count occurrences of an element
count_banana = fruits.count("banana")
print("Count of 'banana':", count_banana)

# 6. Index of an element
index_cherry = fruits.index("cherry")
print("Index of 'cherry':", index_cherry)

# 7. Sort the list
fruits.sort()
print("After Sorting:", fruits)

# 8. Reverse the list
fruits.reverse()
print("After Reversing:", fruits)

# 9. Slice the list
print("Sliced [1:4]:", fruits[1:4])

# 10. Copy the list
fruits_copy = fruits.copy()
print("Copied List:", fruits_copy)

# 11. Extend the list with another list
fruits.extend(["grape", "honeydew"])
print("After Extend:", fruits)

# 12. Clear the list
temp_list = fruits.copy()  # temporary backup
fruits.clear()
print("After Clearing:", fruits)
```

```python
# Demonstrating multiple operations on a tuple

# Create a tuple
my_tuple = (10, 20, 30, 20, 40, 50)
print("Original Tuple:", my_tuple)

# 1. Access elements using index
print("Element at index 2:", my_tuple[2])

# 2. Slicing the tuple
print("Sliced Tuple [1:4]:", my_tuple[1:4])

# 3. Count occurrences of an element
print("Count of 20:", my_tuple.count(20))

# 4. Find index of an element
print("Index of 40:", my_tuple.index(40))

# 5. Length of tuple
print("Length of tuple:", len(my_tuple))

# 6. Check membership
print("Is 30 in tuple?", 30 in my_tuple)

# 7. Iterate through tuple
print("Iterating through tuple:")
for item in my_tuple:
    print(item, end=" ")
print()

# 8. Concatenate two tuples
tuple2 = (60, 70)
combined = my_tuple + tuple2
print("Concatenated Tuple:", combined)

# 9. Nested tuple
nested = (my_tuple, tuple2)
print("Nested Tuple:", nested)

# 10. Convert to list (if you want to modify)
temp_list = list(my_tuple)
temp_list.append(99)
modified_tuple = tuple(temp_list)
print("Modified Tuple (after converting to list):", modified_tuple)
```

```python
# Demonstrating multiple operations on a set

# Create a set
set1 = {1, 2, 3, 4, 5}
print("Original Set:", set1)

# 1. Add a single element
set1.add(6)
print("After add(6):", set1)

# 2. Add multiple elements using update()
set1.update([7, 8, 9])
print("After update([7,8,9]):", set1)

# 3. Remove an element using remove()
set1.remove(2)
print("After remove(2):", set1)

# 4. Remove an element using discard() (no error if not found)
set1.discard(10)  # No error even if 10 is not in set
print("After discard(10):", set1)

# 5. Remove a random element using pop()
popped = set1.pop()
print("Element popped:", popped)
print("After pop():", set1)

# 6. Clear all elements
set_copy = set1.copy()  # save copy before clearing
set1.clear()
print("After clear():", set1)

# 7. Set union
a = {1, 2, 3}
b = {3, 4, 5}
print("Set A:", a)
print("Set B:", b)
print("Union (a | b):", a | b)

# 8. Set intersection
print("Intersection (a & b):", a & b)

# 9. Set difference
print("Difference (a - b):", a - b)

# 10. Symmetric difference
print("Symmetric Difference (a ^ b):", a ^ b)

# 11. Membership test
print("Is 2 in set a?", 2 in a)

# 12. Length of a set
print("Length of set b:", len(b))

# 13. Copying a set
copied_set = b.copy()
print("Copied Set B:", copied_set)
```
