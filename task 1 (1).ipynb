```python
# Demonstrate Multiple File Operations in Python

# 1. Write to a file (create new or overwrite)
with open("example.txt", "w") as file:
    file.write("This is the first line.\n")
    file.write("This is the second line.\n")

# 2. Read from the file
with open("example.txt", "r") as file:
    print("Reading the file after writing:")
    content = file.read()
    print(content)

# 3. Append to the same file
with open("example.txt", "a") as file:
    file.write("This is an appended line.\n")

# 4. Read again after appending
with open("example.txt", "r") as file:
    print("Reading the file after appending:")
    updated_content = file.read()
    print(updated_content)

# 5. Reading line-by-line
with open("example.txt", "r") as file:
    print("Reading line by line:")
    for line in file:
        print(line.strip())

# 6. Using 'with' ensures the file is automatically closed
print("File operations completed.")
```

```python
# Demonstrating multiple operators in Python

# --- Assignment Operators ---
a = 10
b = 5

# --- Arithmetic Operators ---
add = a + b        # Addition
sub = a - b        # Subtraction
mul = a * b        # Multiplication
div = a / b        # Division

print("Arithmetic Operators:")
print("Addition:", add)
print("Subtraction:", sub)
print("Multiplication:", mul)
print("Division:", div)

# --- Bitwise Operators ---
print("\nBitwise Operators:")
print("a & b (AND):", a & b)
print("a | b (OR):", a | b)
print("a ^ b (XOR):", a ^ b)
print("~a (NOT):", ~a)
print("a << 1 (LEFT SHIFT):", a << 1)
print("a >> 1 (RIGHT SHIFT):", a >> 1)

# --- Assignment Operators ---
print("\nAssignment Operators:")
x = 10
x += 5  # x = x + 5
print("x += 5:", x)
x *= 2  # x = x * 2
print("x *= 2:", x)
x -= 3  # x = x - 3
print("x -= 3:", x)
x //= 4 # x = x // 4
print("x //= 4:", x)

# --- Comparison Operators ---
print("\nComparison Operators:")
print("a == b:", a == b)
print("a != b:", a != b)
print("a > b:", a > b)
print("a <= b:", a <= b)

# --- Logical Operators ---
print("\nLogical Operators:")
cond1 = a > 5
cond2 = b < 3
print("cond1 and cond2:", cond1 and cond2)
print("cond1 or cond2:", cond1 or cond2)
print("not cond1:", not cond1)
```

```python
# Demonstrating multiple string operations

text = "  Hello, Python Programming World!  "

print("Original String:", repr(text))

# 1. Remove leading and trailing whitespaces
stripped = text.strip()
print("Stripped String:", repr(stripped))

# 2. Convert to uppercase
print("Uppercase:", stripped.upper())

# 3. Convert to lowercase
print("Lowercase:", stripped.lower())

# 4. Replace a word
replaced = stripped.replace("Python", "Java")
print("Replaced 'Python' with 'Java':", replaced)

# 5. Split string into a list
words = stripped.split()
print("Split into words:", words)

# 6. Join list back to string
joined = " | ".join(words)
print("Joined with '|':", joined)

# 7. Count occurrences of a character
count_o = stripped.count('o')
print("Count of 'o':", count_o)

# 8. Check if string starts with a word
print("Starts with 'Hello':", stripped.startswith("Hello"))

# 9. Check if string ends with a word
print("Ends with 'World!':", stripped.endswith("World!"))

# 10. Find position of a word
position = stripped.find("Programming")
print("Position of 'Programming':", position)

# 11. String length
print("Length of string:", len(stripped))

# 12. Capitalize first character
print("Capitalized:", stripped.capitalize())

# 13. Title case
print("Title Case:", stripped.title())

# 14. Swap case
print("Swap Case:", stripped.swapcase())

# 15. Is string alphabetic
alpha_check = "Hello".isalpha()
print("Is 'Hello' alphabetic?", alpha_check)
```
